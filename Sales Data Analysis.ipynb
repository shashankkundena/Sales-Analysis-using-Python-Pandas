**What products are most often sold together**

```python
df3.head()
```

```python
df6 = df3[df3['Order ID'].duplicated(keep=False)]
```

```python
df6.head(20)
```

```python
df6['Grouped_orders'] = df6.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df6=df6[['Order ID','Grouped_orders']].drop_duplicates()

df6.head()
```

```python
df6
```

```python
from itertools import combinations
from collections import Counter
```

```python
df6['Products'] = df6['Grouped_orders'].str.split(',')

# Generate all possible combinations of 2 products for each order
combinations_list = [list(combinations(products, 2)) for products in df6['Products']]

# Flatten the list of combinations
all_combinations = [combo for sublist in combinations_list for combo in sublist]

# Count the occurrences of each combination
combination_counter = Counter(all_combinations)

# Get the top 10 combinations
top_combinations = combination_counter.most_common(10)

# Display the top 10 combinations and their counts
for i, (combination, count) in enumerate(top_combinations, 1):
    print(f"{i}. Product Combination: {combination}, Count: {count}")
```

**Do certain products sell better in specific locations?**

```python
product_city_sales = df3.pivot_table(index='Product', columns='City', values='Sales', aggfunc='sum', fill_value=0)
product_city_sales_sorted = product_city_sales.sum(axis=1).sort_values(ascending=False) 
product_city_sales_sorted_df = product_city_sales.loc[product_city_sales_sorted.index]

```

```python
product_city_sales_sorted_df 
```

```python
product_city_sales_sorted_df.plot(kind='barh', stacked=True, figsize=(12, 8))
plt.xlabel('Total Sales')
plt.ylabel('Product')
plt.title('Total Sales of Products in Different Cities')
plt.legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left')  # Add legend for cities
plt.show()
```
